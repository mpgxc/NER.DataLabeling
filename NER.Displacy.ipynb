```python
import spacy
import pandas as pd
from spacy import displacy
```

```python
spacy_nlp = spacy.load("en_core_web_trf")
data = pd.read_csv('data/custom/DATA_PROCESSADO.csv')
```

```python
def chunk_class(tag):
    
    """
        Converte as classes Spacy em um modelo customizado...
    """

    entities = {
        "PERSON": "PESSOA",
        "ORG": "ORGANIZAÇÃO",
        "GPE": "LOCALIZAÇÃO",
        "FAC": "LOCALIZAÇÃO",
        "DATE": "DATA",
        "TIME": "HORA",
        "EVENT": "EVENTO"
    }
    
    result = entities.get(tag)
    
    if not result: return 'O'
    return result
```

```python
all_texts = []
all_colors = []

for idx in range(20):
    
    indexes_tags_mp = []
    indexes_tags_mp_colors = []
    
    doc = spacy_nlp(data.text[idx])
 
    for token in doc.ents:

        tmp_obj = {
            "start": token.start_char, 
            "end": token.end_char, 
            "label": chunk_class(token.label_)
        }
        
        tmp_obj_colors = {
            "start": token.start_char, 
            "end": token.end_char, 
            "label": token.label_
        }

        indexes_tags_mp.append(tmp_obj)
        indexes_tags_mp_colors.append(tmp_obj_colors)
    
    obj_text = {
        "text": str(doc),
        "ents": indexes_tags_mp,
        "title": None
    }
    
    obj_text_colors = {
        "text": str(doc),
        "ents": indexes_tags_mp_colors,
        "title": None
    }


    all_texts.append(obj_text)
    all_colors.append(obj_text_colors)
```

```python
displacy.render(all_texts, style="ent", manual=True)
```

```python
displacy.render(all_colors, style="ent", manual=True)
```
