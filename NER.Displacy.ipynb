```python
import spacy
import pandas as pd
from spacy import displacy
```

```python
spacy_nlp = spacy.load("en_core_web_trf")
data = pd.read_csv('DATA_PROCESSADO.csv')
```

```python
def chunk_class(tag):
    
    """
        Converte as classes Spacy em um modelo customizado...
    """
    
    final = tag
    
    if tag == 'PERSON': 
        final = 'PESSOA'
            
    if tag == 'ORG': 
        final = 'ORGANIZAÇÃO'
            
    if tag == 'GPE' or tag == 'FAC': 
        final = 'LOCOLIZAÇÃO'

    if tag == 'DATE':
        final = 'DATA'
        
    if tag == 'TIME':
        final = 'HORA'
    
    if tag == 'EVENT':
        final = 'EVENTO'
        
    return final
```

```python
all_texts = []

for idx in range(20):
    
    indexes_tags_mp = []
    
    doc = spacy_nlp(data.text[idx])
 
    for token in doc.ents:

        tmp_obj = {
            "start": token.start_char, 
            "end": token.end_char, 
            "label": chunk_class(token.label_),
            'color': '#47d'
        }

        indexes_tags_mp.append(tmp_obj)
    
    obj_text = {
        "text": str(doc),
        "ents": indexes_tags_mp,
        "title": None
    }
    


    all_texts.append(obj_text)
```

```python
displacy.render(all_texts, style="ent", manual=True)
```
